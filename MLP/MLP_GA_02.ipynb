```python
X_test['Age']=X_test['Age'].fillna(X_test['Age'].mean())
X_train['Age']=X_train['Age'].fillna(X_train['Age'].mean())

# X_test['Location']=X_test['Location'].fillna("Other")
# X_train['Location']=X_train['Location'].fillna("Other")

# Calculate mode from training set
mode_value = X_train['GameDifficulty'].mode()[0]

# Fill missing values in both train and test
X_train['GameDifficulty'] = X_train['GameDifficulty'].fillna(mode_value)
X_test['GameDifficulty']  = X_test['GameDifficulty'].fillna(mode_value)

X_test['InGamePurchases']=X_test['InGamePurchases'].fillna(0.0)
X_train['InGamePurchases']=X_train['InGamePurchases'].fillna(0.0)
```

```python
# Replace NaN values in Location with "Other"
X_train['Location'] = X_train['Location'].fillna("Other")
X_test['Location']  = X_test['Location'].fillna("Other")
```

```python
print(X_train.isnull().sum())
print(X_test.isnull().sum())
```

```python
X_train=X_train.drop('PlayerID',axis=1)
X_test=X_test.drop('PlayerID',axis=1)
```
