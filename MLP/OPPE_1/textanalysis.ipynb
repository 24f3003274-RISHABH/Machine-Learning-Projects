```python
from sklearn.feature_extraction.text import CountVectorizer
?CountVectorizer
```

```python
corpus=[
  'This is the first document.',
  'This document is the second document.',
  'And this is the third one.',
  'Is this the first document?',
]
# transform the text data into numerical vector using CountVectorizer
vectorizer=CountVectorizer(lowercase=True)
```

```python
# vectorizer.fit(corpus)
# learn
# Now it learn the vocabulary 

# learn and transform
X=vectorizer.fit_transform(corpus)

# 
```

```python
X.shape
```

```python
vectorizer.get_feature_names_out()
# lexicographically sorted unique words in the corpus
```

```python
print(X.toarray())
```

```python
import pandas as pd
df=pd.DataFrame(X.toarray(),columns=vectorizer.get_feature_names_out())
df
```

```python
# Each string in the corpus is represented with the words present by ignoring the word order , This representation is called as 'Bag of Words' representation.

# upper case and lower case are treated as same word by default in CountVectorizer. We can change this behaviour by setting the parameter lowercase=False while creating the CountVectorizer object.
```

```python
# Experiment : INstantiate a Countvectorizer with "lowercase" =false and  observe the difference in the output.

# Important parameters of CountVectorizer
# lowercase : whether to convert all the characters in the text to lower case before tokenization.


vectorizer=CountVectorizer(lowercase=False)
X=vectorizer.fit_transform(corpus)
print(f"Number of feature: {len(vectorizer.get_feature_names_out())}")
df=pd.DataFrame(X.toarray(),columns=vectorizer.get_feature_names_out())
df
```

```python
# we need to puntuation marks to be removed from the text data before creating the bag of words representation. We can do this by setting the parameter "token_pattern" while creating the CountVectorizer object. By default, the token_pattern is set to r'(?u)\b\w\w+\b' which means that only words with 2 or more characters are considered as tokens and punctuation marks are ignored.
# token_pattern :r"(?u)\b\w\w+\b" is the default value for token_pattern parameter in CountVectorizer. It means that only words with 2 or more characters are considered as tokens and punctuation marks are ignored. We can change this behaviour by setting the token_pattern parameter to a different regular expression pattern.
```

```python
vectorizer=CountVectorizer(lowercase=True,token_pattern=r"[a-zA-Z.?]+")
X=vectorizer.fit_transform(corpus)
print(f"Number of feature: {len(vectorizer.get_feature_names_out())}")
df=pd.DataFrame(X.toarray(),columns=vectorizer.get_feature_names_out())
df
# Regular expression pattern r"[a-zA-Z.?]+" means that only words with 1 or more characters (both upper case and lower case) and punctuation marks (dot and question mark) are considered as tokens.
```

```python
# stop_words=['is','the','this']
# Remove the stop words  
vectorizer=CountVectorizer(
                        lowercase=True,
                        token_pattern=r"[a-zA-Z]+",
                        # stop_words='english')
                        stop_words=['is','the','this']
                        )
X=vectorizer.fit_transform(corpus)
print(f"Number of feature: {len(vectorizer.get_feature_names_out())}")
df=pd.DataFrame(X.toarray(),columns=vectorizer.get_feature_names_out())
df
```
